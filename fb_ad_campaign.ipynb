```python
import numpy as np
import pandas as pd
```

```python
df_fb = pd.read_csv("./data.csv")
df_fb.head(5)
```

```python
#calculate total impressions for the age group 30-34

impression_age = 0
for i, j in df_fb[['age', 'impressions']].iloc():
    if i == '30-34':
        impression_age += j
print("Total impressions for the age group 30-34 : ", impression_age)
```

```python
#get total clicks where report_start between dates 19/08/2017 to 22/08/2017 (both inclusive)

clicks_dates = 0
for i, j in df_fb[['reporting_start', 'clicks']].iloc():
    if int(i[0:2])>=19 and int(i[0:2])<=22:
        clicks_dates += j
print("Total clicks where report_start between dates 19/08/2017 to 22/08/2017 : ", clicks_dates)
```

```python
#get all ad_ids for every campaign_id

campaign_ad = {}
for i in df_fb['campaign_id'].iloc():
    campaign_ad[i] = []
    
for i,j in df_fb[['campaign_id','ad_id']].iloc():
    campaign_ad[i].append(j)

print("All ad_ids for every campaign_id", campaign_ad)
```
